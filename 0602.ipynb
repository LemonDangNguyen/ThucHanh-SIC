```python
import os

# Kiểm tra và cài đặt thư viện requests nếu chưa có
try:
    import requests
except ImportError:
    os.system('pip install requests')
    import requests

# URL của file cần tải
url = 'https://github.com/stefannae/SIC-Artificial-Intelligence/raw/main/SIC_AI_Coding_Exercises/SIC_AI_Chapter_07_Coding_Exercises/book_crime_and_punishment.txt'

# Gửi yêu cầu tải file
response = requests.get(url)

# Lưu file vào thư mục hiện tại
with open('book_crime_and_punishment.txt', 'wb') as file:
    file.write(response.content)

print('File đã được tải xuống và lưu thành công!')
```

```python
import re

# Đọc nội dung file vào biến book
with open('book_crime_and_punishment.txt', 'r') as file:
    book = file.readlines()

# Sử dụng regex để tìm các dòng bắt đầu với "This"
matching_lines = [line for line in book if re.match(r"^This", line)]

# In ra các dòng phù hợp
for line in matching_lines:
    print(line.strip())
```
