```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
a = 2000
b = 5000
base = b - a
```

```python
st.uniform.pdf(4500,a,base)
```

```python
st.uniform.ppf(0.5,a,base)
```

```python
st.uniform.ppf(0.01,a,base)
```

```python
st.uniform.ppf(0.99,a,base)
```

```python
st.uniform.cdf(3500,a,base)
```

```python
st.uniform.cdf(3000,a,base)-st.uniform.cdf(2500,a,base)
```

```python
x=np.linspace(1000,6000,1000)
plt.plot(x, st.uniform.pdf(x,a,base),color='orange')
plt.show()
```

```python
mu = 5
sigma = 2
```

```python
st.norm.pdf(0, loc=mu, scale=sigma)
```

```python
st.norm.ppf(0.5, loc=mu, scale=sigma)
```

```python
st.norm.ppf(0.01, loc=mu, scale=sigma)
```

```python
st.norm.ppf(0.99, loc=mu, scale=sigma)
```

```python
st.norm.cdf(5,loc = mu, scale=sigma)
```

```python
st.norm.cdf(7,loc = mu, scale=sigma) - st.norm.cdf(3,loc = mu, scale=sigma)
```

```python
st.norm.cdf(9,loc = mu, scale=sigma) - st.norm.cdf(1,loc = mu, scale=sigma)
```

```python
st.norm.cdf(11,loc = mu, scale=sigma) - st.norm.cdf(-1,loc = mu, scale=sigma)
```

```python
x=np.linspace(-10,+10,1000)
mu=0
sigma=2
plt.plot(x, st.norm.pdf(x,loc=mu,scale=sigma),color='orange')
plt.show()
```

```python
x=np.linspace(-10,+10,1000)
sigma=2
mu_1 = -1
mu_2 = 0
mu_3 = +1
plt.plot(x, st.norm.pdf(x,loc=mu_1,scale=sigma),color='black')
plt.plot(x, st.norm.pdf(x,loc=mu_2,scale=sigma),color='pink')
plt.plot(x, st.norm.pdf(x,loc=mu_3,scale=sigma),color='navy')
plt.show()
```

```python
x=np.linspace(-10,+10,1000)
sigma_1 = 1
sigma_2 = 2
sigma_3 = 3
mu = 0
plt.plot(x, st.norm.pdf(x,loc=mu,scale=sigma_1),color='black')
plt.plot(x, st.norm.pdf(x,loc=mu,scale=sigma_2),color='pink')
plt.plot(x, st.norm.pdf(x,loc=mu,scale=sigma_3),color='navy')
plt.show()
```
